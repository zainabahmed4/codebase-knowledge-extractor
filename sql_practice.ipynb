## SQL Notebook to view codebase extraction data

```python
import pandas as pd
import sqlite3 
```

```python
df = pd.read_json('weather_app_functions.json')
df
```

```python
conn = sqlite3.connect("codebase_knowledge.db") # This connects Python to your database file
df.to_sql("functions", conn, if_exists="replace", index=False) # This writes the DataFrame to a table named "functions" in the database, replacing it if it already exists and not including the DataFrame index as a column.
```

### SQL Queries:

```python
pd.read_sql_query("SELECT * FROM functions;", conn)
```

```python
pd.read_sql_query("SELECT function_name from functions WHERE purpose = 'UI handling';", conn)
```

```python
pd.read_sql_query("SELECT purpose, COUNT(*) FROM functions GROUP BY purpose;", conn)
```
