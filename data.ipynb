## Extracted Data Organization

Using data from the Python Weather Info App

```python
import os
import pandas as pd
import numpy as np

df = pd.read_csv("extracted_functions.csv")

df
```

```python
# Fix some formatting to clean it up:

# Remove the leading "# " from the description column:
df["description"] = df["description"].str.strip("# ")

# Remove the square brackets and single quotes from the parameters column:
df["parameters"] = df["parameters"].str.strip("[]'")

df = df.replace(r'^\s*$', np.nan, regex=True)

df
```

```python
# Add new column with the purpose of each function:
purpose_list = ['API/data fetching', 'user interaction', 'event handling', 'UI handling', 'UI handling', 'UI handling']
df["purpose"] = purpose_list

# Add new column with the file names:
file_name_list = ['weather_app.py', 'weather_app.py', 'weather_app.py', 'weather_app.py', 'weather_app.py', 'weather_app.py']
df["file_name"] = file_name_list

df
```

```python

df.to_json('weather_app_functions.json', orient='records', indent=4)
```
